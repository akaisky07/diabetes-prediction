```python
train=train.drop('Id',axis=1)
test=test.drop('Id',axis=1)
```

```python
X=train.drop('Outcome',axis=1)
y=train['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Importing necessary libraries
from catboost import CatBoostClassifier
import lightgbm as lgb
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initializing classifiers
cat = CatBoostClassifier(verbose=0)
lgbm = lgb.LGBMClassifier()
xg = xgb.XGBClassifier()
rf = RandomForestClassifier()
lr = LogisticRegression()
svm = SVC()

# Fitting the classifiers
cat.fit(X_train, y_train)
lgbm.fit(X_train, y_train)
xg.fit(X_train, y_train)
rf.fit(X_train, y_train)
lr.fit(X_train, y_train)
svm.fit(X_train, y_train)

# Making predictions on test set
cat_preds = cat.predict(X_test)
lgbm_preds = lgbm.predict(X_test)
xg_preds = xg.predict(X_test)
rf_preds = rf.predict(X_test)
lr_preds = lr.predict(X_test)
svm_preds = svm.predict(X_test)

# Calculating accuracies
cat_acc = accuracy_score(y_test, cat_preds)
lgbm_acc = accuracy_score(y_test, lgbm_preds)
xg_acc = accuracy_score(y_test, xg_preds)
rf_acc = accuracy_score(y_test, rf_preds)
lr_acc = accuracy_score(y_test, lr_preds)
svm_acc = accuracy_score(y_test, svm_preds)

# Printing accuracies
print('CatBoost accuracy: {:.2f}%'.format(cat_acc*100))
print('LightGBM accuracy: {:.2f}%'.format(lgbm_acc*100))
print('XGBoost accuracy: {:.2f}%'.format(xg_acc*100))
print('Random Forest accuracy: {:.2f}%'.format(rf_acc*100))
print('Logistic Regression accuracy: {:.2f}%'.format(lr_acc*100))
print('SVM accuracy: {:.2f}%'.format(svm_acc*100))
```
